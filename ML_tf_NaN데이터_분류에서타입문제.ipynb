```python
import pandas as pd
```

```python
ref = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
iris = pd.read_csv(ref)
iris.head()
```

even if 품종 number, it's stillt 범주형(category)데이터
-> onehotencoding


```python
#onehotencoding
encoding = pd.get_dummies(iris)
encoding.head()
```

onehotencoding cannot read number as 범주형데이터
-> change the VAR to 범주형데이터[Linktext]

```python
print(iris.dtypes) #show data type
```

```python
# chagne 품종type into 'catrgory'
iris['품종'] = iris['품종'].astype('category')
```

```python
print(iris.dtypes)
print(iris.head())
```

```python
#onehotencoding again
encoding = pd.get_dummies(iris)
print(encoding.head())
```

# NA 값의 처리

```python
# NA값을 체크해 줍니다
print(iris.tail())
iris.isna().sum()
```

```python
# put mean into NA value
mean = iris['꽃잎폭'].mean()
print(mean)
iris['꽃잎폭'] = iris['꽃잎폭'].fillna(mean)
iris.tail()
```
